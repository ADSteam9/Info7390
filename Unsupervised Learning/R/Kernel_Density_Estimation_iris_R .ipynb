```r
require(graphics)
#install.packages("MASS")
library(MASS)
```

```r
data(iris)
x=iris$Sepal.Width
y=iris$Petal.Width
```

```r
par(mfrow=c(1,2))
```

```r
plot(density(x,xlab=iris$Sepal.Width))
```

```r
plot(density(y,xlab=iris$Petal.Width))
df <- data.frame(x,y)
```

```r
k <- kde2d(df$x, df$y, n=200)
image(k,xlab="Sepal.Width",ylab="Petal.Width",
      main="Two-Dimensional Kernel Density Estimation")
```
